```
def df_col_propagate(df_name, col_name, col_filter):
    #df_name  - a hole dataframe
    #col_name - column_name where the future row header
    #col_filter - select the row which will be propagated based on filter . For example one of unique values for col_name
    #result has to select one row only!


    #select that row only
    header_old = df_name.select('*').where(df_name[col_name]==col_filter)

    header_df = df_name.select('*').where(df_name[col_name]==col_filter)
    #list the column_names 
    columns_list = df_name.schema.names

    for idx, i in enumerate(columns_list):
        #check and select the value in column i
        one_cell = header_df.select(header_df[i]).collect()[0].asDict()[i]

        #if the value in column i is null , except its the very first column
        if one_cell==None and idx!=0:
            idx_prev = idx-1
            column_prev =columns_list[idx_prev] 
            column_current = i 
            header_df = header_df.withColumn(column_current,header_df[column_prev])

        #if the value is not empty
        elif one_cell!=None:
            pass
        #if this is the very first column
        else:
            pass

    


    #add new row
    result_df = df_name.union(header_df)

    #remove old bad headers
    result_df = result_df.exceptAll(header_old)


    return result_df
```

```
def df_rows_merge(df_name, rows_filters_dict):
    # rows_filters_dict = {'filters':[{'prod':'apple'},{'prod':10}]}

    cols = list()

    for idx, filter  in enumerate(rows_filters_dict['filters']):
        key = [key for key in filter.keys()][0]
        cols.append(key)


    qty_filters = len(rows_filters_dict['filters'])

    

    df_separate_rows = df_name.select('*').where('1=2')

    for idx, col in enumerate(cols):
        df_merge_part = df_name.select('*').where(df_name[col]==rows_filters_dict['filters'][idx][col])
        df_separate_rows = df_separate_rows.union(df_merge_part)


    #Creating dataframe which holds joined rows

    values = []
    

    for column_x in df_separate_rows.schema.names: 
        new_row = str()
        for row in range(df_separate_rows.count()):

            row_part = df_separate_rows.select(df_separate_rows[column_x]).collect()[row].asDict()[column_x]
            if row_part == None:
                row_part = ''

            new_row = str(new_row) + str(row_part) + ' '

        values.append(new_row.strip())

    joined_rows_df = spark.createDataFrame([(values)], schema = df_name.schema)

    final_df = df_name.exceptAll(df_separate_rows)
    final_df = final_df.union(joined_rows_df)
    







    return final_df


```

```
def df_all_colls_rename(df_name, cols_dict):
    #'old_name':'new_name'

    for key, value in cols_dict.items():
        cols_dict[key] = value.replace(' ', '_')

    cols = 'SELECT '
    for old_name, new_name in cols_dict.items():
        try:
            #df_name.schema[old_name].name = new_name
            cols = cols + f'{old_name} as {new_name}, '
            
        except:
            pass

    cols = cols[:-2] + ' from myTable'

    #df = df_name.selectExpr(cols)
    sqlContext.registerDataFrameAsTable(df_name, "myTable")
    df = sqlContext.sql(cols)

    return df
```

```
#propagate value vertically

def df_row_propagate(df_name, col_name, filter_for_propag, unique_id_col):
    #df_name = dataframe name
    #col_name = column to propagate
    #filter_for_propag = filter for values which have to be propagated downwards
    #unique_id_col = column of df which defines a uniqness of the row

    
    row_prop = [i[0] for i in df_name.select(df_name[col_name],df_name[unique_id_col]).collect()]
    row_id = [i[1] for i in df_name.select(df_name[col_name],df_name[unique_id_col]).collect()]

    year_def = ''

    for idx, row_value in enumerate(row_prop):
        if filter_for_propag in row_value:
            year_def = row_value
            row_prop[idx] = Row(added_Period = row_prop[idx] , added_id = row_id[idx])
        else:
            row_prop[idx] = Row(added_Period = row_prop[idx] + ' ' + year_def, added_id = row_id[idx])

        
    new_df = spark.createDataFrame(row_prop)
    final_df = df_name.join(new_df, df_name[unique_id_col]==new_df.added_id, 'inner')
    return final_df


```
